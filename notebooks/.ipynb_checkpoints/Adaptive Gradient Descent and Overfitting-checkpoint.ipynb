# Batch

```python
def grad_desc(X,y,beta0,lrate):
    beta = np.array(beta0)
    num_iterations = 1000
    for n in range(num_iterations):
        for k in range(X.shape[1]):
            grads=np.array([2*(y[i]-np.dot(beta,X.loc[i,:]))*X.loc[i,k] for i in range(len(X))])
            if beta[k]>0:
                beta[k] = beta[k] + lrate*(np.sum(grads)/50 - 0.116)
            elif beta[k]<0:
                beta[k] = beta[k] + lrate*(np.sum(grads)/50 + 0.116)
            else:
                beta[k] = beta[k]
    return beta

beta=grad_desc(pd.DataFrame(X_scaled),y,np.array([0.1]*50),0.01)
```

# Stochastic 

```python
from numpy import linalg as LA
def error(X,y,beta):
    n = len(X)
    return (1.0/n)*LA.norm(y - np.dot(X,beta))
```

```python
from numpy import linalg as LA
def r2(X,y,beta):
    n = len(X)
    return 1 - LA.norm(y - np.dot(X,beta))/LA.norm(y - np.mean(y))
```

# Stochastic $L^2$ regularization

```python
def stoch_l2_grad_desc(X_train,y_train,beta0,lrate,num_iterations):
    
    beta = np.array(beta0)
    betas = []
    beta_final=[]
    errors_train=[]
    errors_test=[]

    for n in range(num_iterations):
        for k in range(X_train.shape[1]):
            i = np.random.choice(range(X_train.shape[0]))
            grads=2*(y_train[i]-np.dot(beta,X_train.loc[i,:]))*X_train.loc[i,k]
            if beta[k]>0:
                beta[k] = beta[k] + lrate*(grads - 0.03)
            else:
                beta[k] = beta[k] + lrate*(grads + 0.03)
        betas.append(beta[0])
        errors_train.append(error(X_train,y_train,beta))
        errors_test.append(error(X_test,y_test,beta))
    beta_final.append(betas)
    return betas,beta,errors_train,errors_test

betas_s,beta_s,errors_train_s,errors_test_s=stoch_l2_grad_desc(X_train,y_train,beta0,lrate1,num_iterations)
```

# Adagrad regularization

```python
def adagrad_grad_desc(X_train,y_train,beta0,lrate,num_iterations):
    beta = np.array(beta0)
  
    betas = []
    eps=0
    beta_final=[]
    errors_train=[]
    grad_sum_sq = np.zeros(X_train.shape[1])
    grad_sum_sq = grad_sum_sq + eps
    errors_test=[]
    for n in range(num_iterations):
        for k in range(X_train.shape[1]):
            i = np.random.choice(range(X_train.shape[0]))
            grads=2*(y_train[i]-np.dot(beta,X_train.loc[i,:]))*X_train.loc[i,k]
            grad_sum_sq[k] = grad_sum_sq[k] + grads**2
            if beta[k]>0:
                beta[k] = beta[k] + lrate*((grads)/np.sqrt(grad_sum_sq[k]+1e-8))
            else:
                beta[k] = beta[k] + lrate*((grads)/np.sqrt(grad_sum_sq[k]+1e-8))
        betas.append(beta[0])
        errors_train.append(error(X_train,y_train,beta))
        errors_test.append(error(X_test,y_test,beta))
    beta_final.append(betas)
    return betas,beta,errors_train,errors_test


betas,beta,errors_train,errors_test=adagrad_grad_desc(X_train,y_train,beta0,0.1,num_iterations)
```

```python
plt.figure(figsize=(8,5))
plt.ylabel("OLS Error")
plt.xlabel('Iterations')
plt.plot(range(1000),errors_train,label='train ADG')
plt.plot(range(1000),errors_test,label='test ADG')
plt.legend()
plt.title("SGD (alpha=0.01) vs AGD (alpha=0.1)")
plt.plot(range(1000),errors_train_s,label='train SGD')
plt.plot(range(1000),errors_test_s,label='test SGD')
plt.legend(loc=1)

plt.figure(figsize=(8,5))
plt.title("Convergence to Coefficient")
plt.xlabel("Iterations")
plt.ylabel('Coefficient')
plt.plot(range(1000),betas,label='AGD')
plt.plot(range(1000),betas_s,label='SGD')
plt.legend(loc=1)
```
